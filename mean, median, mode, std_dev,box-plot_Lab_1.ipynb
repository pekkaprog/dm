```python
arr = [20, 2, 7, 1, 34]
mean1=np.mean(arr)
median1=np.median(arr)
print(mean1)
print(median1)
```

```python
#arr2=np.arrange(20)
arr2=[20, 2, 7, 1, 34]
dev=np.std(arr2) #standard deviation
print(arr2)
print(dev)
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(15)


dataSet = np.random.normal(mean1,dev, 200)  #mean standard deviation
print(dataSet)
figure = plt.figure(figsize =(10, 8))
plt.boxplot(dataSet)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
# Load the Titanic dataset
titanic_data = pd.read_csv('titanic.csv')

# Drop missing values in the 'Age' column
titanic_data_cleaned = titanic_data.dropna(subset=['Age'])

# Print the mean, median, and mode of Age
mean_age = titanic_data_cleaned['Age'].mean()
median_age = titanic_data_cleaned['Age'].median()
mode_age = titanic_data_cleaned['Age'].mode().values[0]

print(f'Mean Age: {mean_age:.2f}')
print(f'Median Age: {median_age:.2f}')
print(f'Mode Age: {mode_age:.2f}')

# Create a box plot for Age using plt.boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(titanic_data_cleaned['Age'].values, vert=False)
plt.title('Box Plot of Age in Titanic Dataset')
plt.xlabel('Age')
plt.show()
```
