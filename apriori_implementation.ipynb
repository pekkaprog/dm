```python
import csv
def csv_to_list(file_path):
    data_list = []
    with open(file_path, 'r') as file:
        csv_reader = csv.reader(file)
        for row in csv_reader:
            data_list.append(row)
    return data_list
def load_dataset():
    file_path=r"C:\Users\harsh\Downloads\HARSHIT - Market_Basket_Optimisation(1).csv" 
    dataset = csv_to_list(file_path)
    return dataset


def scan_dataset(dataset, candidates, min_support):
    support_count = {}
    for transaction in dataset:
        for candidate in candidates:
            if candidate.issubset(transaction):
                if candidate not in support_count:
                    support_count[candidate] = 1
                else:
                    support_count[candidate] += 1
    num_items = len(dataset)
    frequent_items = []
    support_data = {}
    for candidate in support_count:
        support = support_count[candidate] / num_items
        if support >= min_support:
            frequent_items.insert(0, candidate)
        support_data[candidate] = support
    return frequent_items, support_data

def apriori(dataset, min_support=0.5):
    candidates = create_candidates(dataset)
    frequent_items1, support_data = scan_dataset(dataset, candidates, min_support)
    frequent_items = [frequent_items1]
    k = 2
    while len(frequent_items[k - 2]) > 0:
        candidates = generate_candidates(frequent_items[k - 2], k)
        frequent_items_k, support_k = scan_dataset(dataset, candidates, min_support)
        support_data.update(support_k)
        frequent_items.append(frequent_items_k)
        k += 1
    return frequent_items, support_data

def generate_candidates(frequent_items, k):
    candidates = []
    len_frequent_items = len(frequent_items)
    for i in range(len_frequent_items):
        for j in range(i + 1, len_frequent_items):
            L1 = list(frequent_items[i])[:k - 2]
            L2 = list(frequent_items[j])[:k - 2]
            L1.sort()
            L2.sort()
            if L1 == L2:
                candidates.append(frequent_items[i] | frequent_items[j])
    return candidates

if __name__ == "__main__":
    dataset = load_dataset()
    min_support = 0.5
    frequent_items, support_data = apriori(dataset, min_support)
    print("Frequent Items:")
    print(frequent_items)
    print("\nSupport Data:")
#     print(support_data)
    for key, value in support_data.items():
        print(key, ':', value,end='\n')
```
